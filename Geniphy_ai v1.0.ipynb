## Text to Image Generation Using SDXL and Refiner

```python
import gradio as gr
from PIL import Image
import os
import torch
from transformers import pipeline
from diffusers import DiffusionPipeline

###############################
# Image Generation SDXL Model
###############################

def load_model_image_sdxl(model_path, load_model_refiner_sdxl=None, torch_dtype=torch.float16, use_safetensors=True):
    device = "cuda" if torch.cuda.is_available() else "cpu"

    # Base Model
    base = DiffusionPipeline.from_pretrained(model_path,
                                             torch_dtype=torch_dtype,
                                             use_safetensors=use_safetensors)
    base.enable_model_cpu_offload()

    # Refiner Model
    if load_model_refiner_sdxl:
        refiner = DiffusionPipeline.from_pretrained(load_model_refiner_sdxl,
                                                    text_encoder_2=base.text_encoder_2,
                                                    vae=base.vae,
                                                    torch_dtype=torch_dtype,
                                                    use_safetensors=use_safetensors)

        refiner.enable_model_cpu_offload()
        return base, refiner

    else:
        return base, None


def generate_image_sdxl(base, prompt, refiner=None, num_inference_steps=50, guidance_scale=4, high_noise_frac=0.6,
                        output_type="latent", verbose=False, temperature=0.65):

    if refiner:
        image = base(prompt=prompt,
                     num_inference_steps=num_inference_steps,
                     denoising_end=24,
                     output_type=output_type,
                     verbose=verbose,
                     temperature=temperature,
                     guidance_scale=guidance_scale).images

        image = refiner(prompt=prompt,
                        num_inference_steps=num_inference_steps,
                        denoising_end=16,
                        image=image,
                        verbose=verbose).images[0]

    else:
        image = base(prompt=prompt,
                     num_inference_steps=num_inference_steps,
                     guidance_scale=guidance_scale).images[0]

    return image


# Model paths
model_path_sdxl = "/content/Models/models--stabilityai--stable-diffusion-xl-base-1.0/snapshots/462165984030d82259a11f4367a4eed129e94a7b"
model_path_sdxl_refiner = "/content/Models/models--stabilityai--stable-diffusion-xl-refiner-1.0/snapshots/5d4cfe854c9a9a87939ff3653551c2b3c99a4356"

# Load models
base, refiner = load_model_image_sdxl(model_path_sdxl, model_path_sdxl_refiner)


def generate_images(prompt, num_images):
    images = []
    for _ in range(num_images):
        image = generate_image_sdxl(base, prompt, refiner)
        images.append(image)
    return images


# Gradio interface
prompt_input = gr.Textbox(label="Enter your prompt", placeholder="Enter your prompt here")
num_images_slider = gr.Slider(minimum=1, maximum=10, step=1, label="Number of Images", value=2)
output_gallery = gr.Gallery(label="Generated Images")

gr.Interface(
    fn=generate_images,
    inputs=[prompt_input, num_images_slider],
    outputs=output_gallery,
    title="Geniphy.ai",
    description="Generate images based on your prompt using a diffusion model"
).launch()
```
